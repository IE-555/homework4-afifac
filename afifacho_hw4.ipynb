```python
dt = pd.read_csv('student_grades.csv')

A = dt[(dt[' avgScore '] >= 90) & (dt[' avgScore '] < 100)]
B = dt[(dt[' avgScore '] >= 80) & (dt[' avgScore '] < 90)]
C = dt[(dt[' avgScore '] >= 70) & (dt[' avgScore '] < 80)]
D = dt[(dt[' avgScore '] >= 60) & (dt[' avgScore '] < 70)]
F = dt[(dt[' avgScore '] < 60)]

A['studentID'] = 'A'
B['studentID'] = 'B'
C['studentID'] = 'C'
D['studentID'] = 'D'
F['studentID'] = 'F'

grades = pd.concat([A,B,C,D,F])
histdata = grades['studentID'].value_counts()
histdata_real = histdata.sort_index(ascending=False)
print(histdata_real)

plot = plt.bar(histdata_real.index, histdata_real, edgecolor = 'black', color = 'orange')
plt.xlabel('Grade')
plt.ylabel('Count')
plt.title('Grade Distribution')

for bar in plot:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, str(int(height)), ha='center', va='bottom')

plt.show()
```

### Data based on climate change trend is the same in Toronto.
- Use Correlation Plot

- **URL of Example Code**:  https://seaborn.pydata.org/examples/spreadsheet_heatmap.html 
- **URL of Sample Data**:  https://climate.weather.gc.ca 
- **Dataset**: en_climate_monthly_Toronto.csv 

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.animation as animation 
import pandas

df_month = pandas.read_csv('en_climate_monthly_Toronto.csv')

# Group DataFrame by Year
df_year = df_month.groupby('Year').aggregate({'Mean Temp (Â°C)': 'mean', 
        'Extr Max Temp (Â°C)': 'max','Extr Min Temp (Â°C)': 'min', 
        'Total Rain (mm)': 'sum','Total Snow (cm)': 'sum'})

# Convert the Mean Temp from Celsius to Fahrenheit 
df_year['Mean Temp (Â°F)'] = (df_year['Mean Temp (Â°C)']*9/5)+32

# Reset the index
df_year = df_year.reset_index()
```

```python
# Pairwise correlation of columns.
corr_matrix_y = df_year.corr()

# Matrix Plot
fig1, ax1 = plt.subplots(figsize=(12,9)) 
sns.heatmap( corr_matrix_y, ax = ax1, annot=True,  
            cmap='Blues').set(title='Correlation Matrix')
```

# NOTE: Correlation Plot

Correlation falls between -1 and 1. 

100% mean its fully correlated: You can see one 100% correlated in the Mean Temp(C) and Mean Temp(F).

As the year increases so does the monthly mean temperature in Celsius on the left of the correlation plot
Positive correlation values mean a positive trend.

# Extra Plot

```python
# Multiple Plots
fig2, ax2 = plt.subplots(2, 2, figsize=(14,10))
plt.suptitle('Snowfall trends in Toronto', fontsize=14, fontweight='bold')

# Filter rows with more than 2cm monthly snowfall
df_snow = df_month[(df_month['Total Snow (cm)'] > 2)]

# Create a Histogram
sns.histplot(ax=ax2[0,0],data=df_snow['Mean Temp (Â°C)'], bins=9, 
  kde = True).set(title='Snowfall Frequency based on Monthly Mean Temperature')

# Create a Scatter plot 

sns.scatterplot(ax=ax2[0,1],x='Total Snow (cm)', y='Mean Temp (Â°C)', 
          data=df_snow,color='blue',alpha=0.8).set(
              title='Relationship between Snowfall and Monthly Mean Temp.')

ax2[0,1].text(25, -12.5, "Outlier", color = 'r')
ax2[0,1].text(110, -4, "Outlier", color = 'r')


# Create a Bar plot
sns.barplot(ax=ax2[1,0],orient = 'h', x = df_snow['Total Snow (cm)'],
            y = df_snow['Month'],color='royalblue').set(
                title='Monthly Mean Snowfall')
ax2[1,0].set_xlabel('Mean Snow fall(cm)')
ax2[1,0].set_yticklabels(['Jan','Feb','Mar','Apr','May','Oct','Nov','Dec'])

# Create a boxplot
sns.boxplot(ax=ax2[1,1],x='Month', y='Total Snow (cm)', 
            data=df_month,color='lightblue').set(
                title='Monthly Distribution of Snowfall')
ax2[1,1].set_xticklabels(['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug',
                          'Sep','Oct','Nov','Dec'])
```

## Graph 1: Snowfall Frequency based on Monthly Mean Temperature
- Snowfall occurs most frequently  in -3 and -4 degrees celsius temperature.

## Graph 2: Relationship between Snowfall and Monthly Mean Temp
- Most of the snowfall occurred between 0 & -5 deg.C.
- an outlier occurrs when the temp is around -12 deg.
- no snowfall when the monthly mean temperature goes below -10 deg.C. 

## Graph 3: Monthly mean snowfall
- We see mean snowfall of around 38 cm in January and close to 30 cm in December.

## Graph 4: Monthly Distribution of Snowfall
- display data from all months, including the months that had less than 2 cm snowfall.
